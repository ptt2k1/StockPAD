### 4. Lưu dữ liệu

### 5. Biểu diễn

```r
install.packages('repr')
library(repr)
options(repr.plot.width=25, repr.plot.height=8)
price_df <- loadPrice('MWG','FPT', 'VCB','BID', 'VN30') %>%
  filter(ticker %in% c('FPT')) %>%
  group_by(ticker) %>%
    ggplot(aes(x=date, y=close, color = ticker)) +
    geom_line()+
    ggtitle('Price Data FPT')
    price_df
```

### 6. Biểu Diễn Giá 04 Cổ Phiếu từ 01/01/2015

```r
price_df <- loadPrice('MWG','FPT', 'VCB','BID', 'VN30') %>%
  filter(ticker %in% c('MWG','FPT', 'VCB','BID'),
        date >= '2015-01-01') %>%
  group_by(ticker) %>%
    ggplot(aes(x=date, y=close, color=ticker)) +
    geom_line()+
    ggtitle('Price Data of 04 Stocks from 01/01/2015')
    price_df
```
